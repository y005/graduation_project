```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium.webdriver.common.keys import Keys
import time

def titleExtract(driver,keyword):
    
    url = 'https://www.youtube.com/results?search_query={}'.format(keyword)

    driver.get(url)
    soup = bs(driver.page_source, 'html.parser')
    
    name = soup.select('a#video-title')
    video_url = soup.select('a#video-title')
    view = soup.select('a#video-title')

    if(len(name)==0):
        return ["","",""]
    
    name_list = []
    url_list = []
    view_list = []

    idx=0
    for i in range(len(name)):
        if idx> 0 : 
            break
        name_list.append(name[i].text.strip())
        view_list.append(view[i].get('aria-label').split()[-1])
        idx = idx+1
    
    idx=0
    for i in video_url:
        if idx > 0:
            break
        url_list.append('{}{}'.format('https://www.youtube.com',i.get('href')))
        idx = idx+1
        
    return [name_list[0],url_list[0],view_list[0]]
```

```python
youtube1 = pd.read_csv('네이버.csv')
youtube2 = pd.read_csv('카카오.csv')
youtube3 = pd.read_csv('애플.csv')
youtube4 = pd.read_csv('테슬라.csv')

youtube1['Title'] = ""
youtube2['Title'] = ""
youtube3['Title'] = ""
youtube4['Title'] = ""
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
    

for i in range(len(youtube4)):
    result = titleExtract(driver,youtube4['Video ID'][i])
    youtube4['Title'][i] = result[0]

driver.close()
```
